```
display(
  df
    .select(df["OrderID"],df["Rate"]*df["Quantity"])
    .groupBy("OrderID")
    .sum()
    .limit(10)
)
```

```
%sql

SELECT OrderID, SUM(Rate * Quantity) AS OrderTotal
FROM `CogsleyServices_SalesData_US_csv`
GROUP BY OrderID
LIMIT 10
```
